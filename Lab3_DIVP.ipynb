# Grey Level Slicing

# gamma factor transformation / Log Transformations

```python
#log transformation
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
xray = cv2.imread('/content/drive/MyDrive/SEM 6/DIVP LAB/LAb3/xray.jpg', 1)
print("Original X-Ray Image\n")
cv2_imshow(xray)

c= 255 / (np.log(1 + np.max(xray)))
log_image = c * np.log(1 + xray)
print("\n")
print(" X-Ray Image After log transformation\n")

cv2_imshow(log_image)
```

```python

# gamma transformation
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
xray = cv2.imread('/content/drive/MyDrive/SEM 6/DIVP LAB/LAb3/xray.jpg', 1)
print("Original X-Ray Image\n")
cv2_imshow(xray)

for gamma in [0.1, 0.5 ,1.2, 2.2]:
  # normalize the img
  for i in range (4):
    gamma_corrected = np.array(255*(xray / 255) ** gamma, dtype = 'uint8')

  #display corrected image
  print(f'gamma = {gamma}')

  print("\n")
  cv2_imshow(gamma_corrected)

```

# Bit Place Slicing

```python


# Bit Place Slicing
img = cv2.imread('/content/drive/MyDrive/SEM 6/DIVP LAB/LAb3/xray.jpg', 0)
print("Original Image\n")
cv2_imshow(img)

for i in range(8):
    bit_plane = np.zeros(img.shape, np.uint8)
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if (img[x, y] & (1 << i)):
                bit_plane[x, y] = 255

    print(f"Bit Plane {i}")
    cv2_imshow(bit_plane)
```

# Thresholding

```python

# Thresholding

img = cv2.imread('/content/drive/MyDrive/SEM 6/DIVP LAB/LAb3/xray.jpg', 0)
print("Original Image\n")
cv2_imshow(img)

ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
print("Binary Threshold")
cv2_imshow(thresh1)

ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
print("Binary Inverse Threshold")
cv2_imshow(thresh2)

ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
print("Truncated Threshold")
cv2_imshow(thresh3)

ret, thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
print("Threshold to Zero")
cv2_imshow(thresh4)

ret, thresh5 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO_INV)
print("Threshold to Zero Inverse")
cv2_imshow(thresh5)
```

Contranst Stretching

```python


import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/SEM 6/DIVP LAB/LAb3/xray.jpg', 0)

print("Original Image")
cv2_imshow(img)


min_val = np.min(img)
max_val = np.max(img)

a = 0
b = 255

stretched_img = np.zeros_like(img, dtype=np.uint8)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    stretched_img[i,j] = ((img[i,j] - min_val) / (max_val - min_val)) * (b-a) + a

print("Contrast Stretched Image")
cv2_imshow(stretched_img)
```
