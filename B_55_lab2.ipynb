Multiplication of images

```python
mul1=cv2.imread("/content/drive/MyDrive/DIVP LAB/Lab2/cameraman.jpg")
print("img 1")
cv2_imshow(mul1)
# multiply
mul =cv2.multiply(mul1,2)
print("img after mul by 2")
cv2_imshow(mul)

mul =cv2.multiply(mul1,4)
print("img after mul by 4")
cv2_imshow(mul)
```

```python
div1=cv2.imread("/content/drive/MyDrive/DIVP LAB/Lab2/cameraman.jpg")
print("img 1")
cv2_imshow(div1)
# multiply
div =cv2.divide(div1,2)
print("img after div by 2")
cv2_imshow(div)

div =cv2.divide(div1,4)
print("img after div by 4")
cv2_imshow(div)
```

```python
not1 = cv2.imread("/content/drive/MyDrive/DIVP LAB/Lab2/NOT operation.jpg")
not1 = cv2.resize(not1, (400, 400))
print("image ")
cv2_imshow(not1)
print("Not operator on image")
not2=cv2.bitwise_not(not1)
cv2_imshow(not2)
```

```python
img1 = cv2.imread("/content/drive/MyDrive/DIVP LAB/Lab2/Logical1.jpg")
img11 = cv2.resize(img1, (200, 200))
print("image 1\n")
cv2_imshow(img11)

img2 = cv2.imread("/content/drive/MyDrive/DIVP LAB/Lab2/logical2.jpg")
img = cv2.resize(img2, (200, 200))
print("image 2\n")
cv2_imshow(img)

print(" logical And operator on image\n")
ans=cv2.bitwise_and(img11,img)
cv2_imshow(ans)
```
